```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Book.csv')
```

```python
df.head()
```

```python
df
```

```python
hours = np.array(df['hours'].values)
results = np.array(df['results'].values)
```

```python
plt.scatter(hours, results, color='orange')
plt.xlabel("hours")
plt.show()
```

```python
m,b = np.polyfit(hours,results,1)
plt.xlabel("Hours")
plt.plot(hours, results, 'o', color='orange')
plt.plot(hours, m*hours+b)
```

```python
x_data = df[['hours']]
y_data = df[['results']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

model.predict(x_test)

```python
model.predict(x_test)
```

```python
model.predict([[6], [15], [19], [25]])
```

```python
model.score(x_test, y_test)
```
